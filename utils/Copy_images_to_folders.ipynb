Rename folder

```python
import os

ori_path = r"C:\Users\tjian\Desktop\TUD-GV Dataset\TUD-GV"

folders = os.listdir(ori_path)

for folder in folders: 
    dir = ori_path + '/' + str(folder)
    sub_folders = os.listdir(dir)
    for sub_folder in sub_folders:
      if sub_folder == "No litter":
         sub_folder_path = dir + '/' + str(sub_folder)
         sub_folder_path_new = dir + '/' + "0 - No litter"
         os.rename(sub_folder_path, sub_folder_path_new)
      if sub_folder == "Little litter":
         sub_folder_path = dir + '/' + str(sub_folder)
         sub_folder_path_new = dir + '/' + "1 - Little litter"
         os.rename(sub_folder_path, sub_folder_path_new)
      if sub_folder == "Moderate litter":
         sub_folder_path = dir + '/' + str(sub_folder)
         sub_folder_path_new = dir + '/' + "2 - Moderate litter"
         os.rename(sub_folder_path, sub_folder_path_new)
      if sub_folder == "Lots of litter":
         sub_folder_path = dir + '/' + str(sub_folder)
         sub_folder_path_new = dir + '/' + "3 - Lots of litter"
         os.rename(sub_folder_path, sub_folder_path_new)
```
